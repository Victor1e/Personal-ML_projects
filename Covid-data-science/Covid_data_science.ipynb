```python
covid=df2
```

```python
covid['Active_cases']=covid['cases']-(covid['deaths']+covid['recovered'])
```

```python
covid.tail()
```

```python
covid['Recovery_rate']=covid['recovered']*100/covid['cases']
covid.tail()
```

```python
covid['Recovery_rate'].mean()
```

```python
plt.plot(covid['Recovery_rate'])
```

```python
covid['Mortality_rate']=covid['deaths']*100/covid['cases']
covid.tail()
```

```python
covid['Mortality_rate'].mean()
```

```python
plt.plot(covid['Mortality_rate'])
```
